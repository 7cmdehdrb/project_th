```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
file_path = "/home/min/7cmdehdrb/project_th/rtde_log.csv"

data = pd.read_csv(file_path)

# display(data.keys())

R_elbow = data["R_elbow_p"].values
S_elbow = data["S_elbow_p"].values

# R_elbow = data["R_shoulder_lift_p"].values
# S_elbow = data["S_shoulder_lift_p"].values

d_elbow = R_elbow - S_elbow

abs_d_elbow = np.abs(d_elbow)

d_elbow_max = np.rad2deg(abs_d_elbow.max())
d_elbow_min = np.rad2deg(abs_d_elbow.min())
d_elbow_mean = np.rad2deg(abs_d_elbow.mean())
d_elbow_std = np.rad2deg(abs_d_elbow.std())

display(d_elbow_max)
display(d_elbow_min)
display(d_elbow_mean)
display(d_elbow_std)

df = pd.DataFrame(
    {
        "d_elbow": d_elbow,
    }
)

display(df)
```
